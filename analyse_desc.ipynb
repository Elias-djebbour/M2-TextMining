1. Importation des bibliothèques nécessaires

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from collections import Counter
import string
```

2. Chargement des données

```python
path = "./archive/medium_articles.csv"

df = pd.read_csv(path, delimiter=",", quotechar='"')
```

```python
print(df['tags'].head())
# print(df.head())
```

3. Analyse descriptive de base

```python
# # Informations générales sur le dataset
# print(df.info())

# # Statistiques descriptives pour les colonnes numériques (s'il y en a)
# print(df.describe())

# Nombre d'articles par auteur
author_counts = df['authors'].explode().value_counts()
print(author_counts)

# Nombre d'articles par tag
tag_counts = df['tags'].explode().value_counts()
print(tag_counts)
```

```python
# Supposons que df est déjà chargé et que la colonne 'tags' est déjà nettoyée
# Exploser la colonne 'tags' pour avoir un tag par ligne
tags_exploded = df['tags'].explode()

# Compter la fréquence de chaque tag
tag_counts = tags_exploded.value_counts()

# Sélectionner les 20 tags les plus fréquents
top_20_tags = tag_counts.head(20)

# Visualiser les 20 tags les plus fréquents
plt.figure(figsize=(10, 6))
sns.barplot(x=top_20_tags.values, y=top_20_tags.index, palette='viridis')
plt.title('Top 20 des tags les plus fréquents')
plt.xlabel('Nombre d\'articles')
plt.ylabel('Tags')
plt.show()
```
